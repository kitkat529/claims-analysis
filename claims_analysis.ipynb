# **RELATIONAL DATA ANALYSIS**

```python
# 1. Who are the top 5 billing providers by number of claims?:  SB Internists, SB Surgical Associates
  #  New York Spine and Brain Surgery, Univ. Assoc. In OB/GYN, SB Psychiatric Associates


## Group by BillinProvFirstName & BillingProviderNPI then count number of claims
top_billing_providers = df_header.groupby(['BillingProvFirstName', 'BillingProviderNPI']).size().reset_index(name='ClaimsNumber')

## Sort "ClaimsNumber" in descending order & get top 5
top_5_billing_providers = top_billing_providers.sort_values(by='ClaimsNumber', ascending=False).head(5)

print("Top 5 Billing Providers by Number of Claims:")
display(top_5_billing_providers)

## Bar Chart:

plt.figure(figsize=(10, 6))
sns.barplot(x='ClaimsNumber', y='BillingProvFirstName', data=top_5_billing_providers, palette='viridis')
plt.title('Top 5 Billing Providers by Number of Claims')
plt.xlabel('Number of Claims')
plt.ylabel('Provider Name')
plt.tight_layout()
plt.show()
```

```python
# 2. What are the top 5 primary payers by claim volume?
    # Calculate percentage of total claims per year:

# Get claim volume for primary payer using value_counts() since data is in single column
payer_claim_counts = df_header['PrimaryPayerName'].value_counts()

#value_counts() automatically sorts in descending order
top_5_payers = payer_claim_counts.head(5)

# Get total number of claims
total_claims = df_header.shape[0]  ## This returns the number of rows in the df_header dataset
                                   ## The number of columns would be ".shape[1]" since
                                   ## .shape = (rows, columns)

# Get percentage of total claims for each payer
top_5_payers_percentage = (top_5_payers / total_claims) * 100

## Combine info into DataFram for easier visualization
top_payers_df = pd.DataFrame({
    'Payer Name': top_5_payers.index,
    'Claim Count': top_5_payers.values,
    'Percentage': top_5_payers_percentage.values
})

print("Top 5 Primary Payers by Claim Volume and Percentage of Total Claims:")
display(top_payers_df)

# Pie chart of results

plt.figure(figsize=(10, 10))
plt.pie(top_payers_df['Percentage'], labels=top_payers_df['Payer Name'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis'))
plt.title('Percentage Distribution of Top 5 Primary Payers')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

```python
# 3. What are the 10 most frequently appearing diagnostic codes:

## J96.01 - Acute respiratory failure with hypoxia
## I10 - Essential (primary) hypertension
## E78.5  Hyperlipidemia, unspecified
## G93.5  Compression of brain
## D64.9  Anemia, unspecified
## I25.10  Atherosclerotic heart disease of native coronary artery without angina pectoris
## I61.9  Intracerebral hemorrhage, unspecified
## I48.91  Unspecified atrial fibrillation
## I60.8  Other nontraumatic subarachnoid hemorrhage
## I50.9  Heart failure, unspecified

diagnostic_code_counts = df_code['CodeValue'].value_counts()
top_10_diagnositc_codes = diagnostic_code_counts.head(10)

print("Top 10 Most Frequently Appearing Diagnostic Codes:")
display(top_10_diagnositc_codes)
```

```python
# 4. What are the 10 most frequently billed procedure codes?

procedure_code_counts = df_line['HCPCS'].value_counts()
top_10_procedure_codes = procedure_code_counts.head(10)

print("Top 10 Most Frequently Billed Procedure Codes: HCPCS Code and Frequency Count")
display(top_10_procedure_codes)

## Bar Chart

plt.figure(figsize=(12, 7))
sns.barplot(x=top_10_procedure_codes.values, y=top_10_procedure_codes.index, palette='viridis')
plt.title('Top 10 Most Frequently Billed Procedure Codes')
plt.xlabel('Frequency Count')
plt.ylabel('HCPCS Code')
plt.tight_layout()
plt.show()
```

```python
# 5. How many claims were submitted for each PlaceOfService?
     # What percent are 'INPATIENT' vs. 'DOCTOR'S OFFICE'?  INPATIENT = 59.54   DOCTOR'S OFFICE = 34.02

Place_of_service_claims = df_header['PlaceOfService'].value_counts()

print("Number of claims submitted for each PlaceOfService:")
display(Place_of_service_claims)


## Percentage of "INPATIENT" vs. "DOCTOR'S OFFICE"

total_claims = len(df_header)   ## can also use df_header.shape[0] as before

# Inpatient percentage:
inpatient_claims = df_header[df_header['Facility_Location'] == 'INPATIENT'].shape[0]
percentage_inpatient = (inpatient_claims / total_claims) * 100


# Doctor's office percentage:
doctors_office_claims = df_header[df_header['Facility_Location'] == 'DOCTOR\'S OFFICE'].shape[0]
percentage_doctors_office = (doctors_office_claims / total_claims) * 100


print(f"\nTotal Claims: {total_claims}")
print(f"Claims for INPATIENT (from Facility_Location): {inpatient_claims} ({percentage_inpatient:.2f}%) ")
print(f"Claims for DOCTOR'S OFFICE (from Facility_Location): {doctors_office_claims} ({percentage_doctors_office:.2f}%)")
```

# **ADVANCED ANALYSIS WITH JOINS**

```python
# Merge df_header and df_line. ProspectiveClaimId = common key.

df_header_line_merge = pd.merge(df_header, df_line, on='ProspectiveClaimId', how='inner')
```

```python
df_header_line_merge.columns
```

```python
# 6. Claims with high service line counts

# Find total service lines
service_lines_per_claim = df_header_line_merge.groupby('ProspectiveClaimId').size().reset_index(name='ServiceLines')

# Find claims with 5+ service lines
claims_with_5_or_more_service_lines = service_lines_per_claim[service_lines_per_claim['ServiceLines'] >= 5]

# Find total charges per claim
total_charges_per_claim = df_header_line_merge.groupby('ProspectiveClaimId')['Charges'].sum().reset_index(name='TotalCharges')

# Merge the two newly created df
billing_df = pd.merge(claims_with_5_or_more_service_lines, total_charges_per_claim, on='ProspectiveClaimId', how='inner')

# Merge new df with df_header to get BillingProvFirstName column
billing_total_df = pd.merge(billing_df, df_header, on= 'ProspectiveClaimId', how='inner')

# Select and rename desired columns into new df
billing_summary = billing_total_df[['ProspectiveClaimId','BillingProvFirstName', 'ServiceLines', 'TotalCharges']]
billing_summary = billing_summary.rename(columns={'ProspectiveClaimId': 'ClaimId', 'BillingProvFirstName': 'Provider Name', 'ServiceLines': 'Number of Lines', 'TotalCharges': 'Total Charges'})
print(f'Total Number of Service Lines and Charges by Provider:')
display(billing_summary)
```

```python
# Merge df_line and df_code. Common key = ProspectiveClaimId

df_line_code_merge = pd.merge(df_line, df_code, on= 'ProspectiveClaimId', how= 'inner')
```

```python
# 7. Find the most common diagnosis code (CodeValue) associated with CPT code 99291
      # J96.01 - Acute respiratory failure with hypoxia

# Filter for CPT code 9921
cpt_99291 = df_line_code_merge[df_line_code_merge['HCPCS'] == '99291']
display(cpt_99291.head())

# Find most common diagnosis code (CodeValue) in cpt_99291
most_common_diagnosis = cpt_99291['CodeValue'].value_counts().idxmax()

print(f"Most common diagnosis code with CPT code 99291: {most_common_diagnosis}")
```

```python
# 8. Charges by payer: Calculate total charges per claim

# Merge df_header and df_line
df_header_line_merge = pd.merge(df_header, df_line, on= 'ProspectiveClaimId', how= 'inner')

# Group by PrimaryPayerName for total charges
total_charges_per_payer = df_header_line_merge.groupby('PrimaryPayerName')['Charges'].sum().reset_index(name= 'TotalChargesPerPayer')

# Group by PrimaryPayerName for number of claims
num_claims_per_payer = df_header_line_merge.groupby('PrimaryPayerName')['ProspectiveClaimId'].nunique().reset_index(name= 'NumClaimsPerPayer')

# New df with charges and claims per payer
payer_summary = pd.merge(total_charges_per_payer, num_claims_per_payer, on= 'PrimaryPayerName', how= 'inner')

# Find average charges per claim
payer_summary['AvgChargesPerClaim'] = payer_summary['TotalChargesPerPayer'] / payer_summary['NumClaimsPerPayer']

# Sort by total charges and display to 10 payers
payer_summary = payer_summary.sort_values(by= 'TotalChargesPerPayer', ascending= False).head(10)

print(f'Top 10 Total Charges by Payer')
display(payer_summary)
```

# **CREATIVE ANALYSIS**

```python
# Which providers bill for the most complex cases?

# Find unique icd10 codes in df_code
unique_icd10_codes_per_claim = df_code.groupby('ProspectiveClaimId')['CodeValue'].nunique().reset_index(name='UniqueICD10Codes')

# Merge new df with df_header
unique_icd10_df = pd.merge(unique_icd10_codes_per_claim, df_header, on= 'ProspectiveClaimId', how= 'inner')

# Group by BillingProvFirstName & get avereage unique ICD10 codes
avg_complex_cases_per_provider = unique_icd10_df.groupby('BillingProvFirstName')['UniqueICD10Codes'].mean()

# Get value counts & sort by descending order based on average complex cases
avg_complex_cases_per_provider = avg_complex_cases_per_provider.sort_values(ascending= False)
print(f'Most Complex Cases by Provider:')
display (avg_complex_cases_per_provider)
```

```python
# Creative Ananlysis Visualization

plt.figure(figsize=(12, 7))
sns.barplot(y=avg_complex_cases_per_provider.index, x=avg_complex_cases_per_provider.values, palette='viridis')
plt.title('Average Number of Unique ICD10 Codes per Provider')
plt.xlabel('Average Unique ICD10 Codes')
plt.ylabel('Billing Provider')
plt.tight_layout()
plt.show()
```
